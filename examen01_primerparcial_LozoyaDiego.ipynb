### Opciones de KPI:

### **Opción 1: Tasa de Sesiones por Cliente**
   - *Definición*: Número promedio de sesiones supervisadas por cliente.
   - *Fórmula*: $
     \text{Tasa de Sesiones por Cliente} = \frac{\text{Número de Sesiones por Semana}}{\text{Clientes Existentes}}$

### **Opción 2: Tasa de Retorno de Inversión en Marketing (ROI de Marketing)**
   - *Definición*: Mide el retorno generado por los gastos de marketing en relación con los ingresos obtenidos.
   - *Fórmula*: $\text{ROI de Marketing} = \frac{\text{Ingresos Totales} - \text{Gastos de Marketing}}{\text{Gastos de Marketing}}$

### **Opción 3: Tasa de Crecimiento de Nuevos Clientes**
   - *Definición*: El porcentaje de aumento o disminución en la cantidad de nuevos clientes adquiridos.
   - *Fórmula*: $\text{Tasa de Crecimiento de Nuevos Clientes} = \frac{\text{Nuevos Clientes}_{\text{mes actual}} - \text{Nuevos Clientes}_{\text{mes anterior}}}{\text{Nuevos Clientes}_{\text{mes anterior}}}$

### **Opción 4: Tasa de Churn**
   - *Definición*: El porcentaje de clientes que cancelan el servicio durante un período de tiempo determinado.
   - *Fórmula*: $\text{Tasa de Churn} = \frac{\text{Número de Clientes que Cancelaron}}{\text{Número Total de Clientes Activos al Inicio del Periodo}}$

##**Sección 1**

```python
data = pd.read_csv(data_path + 'gabu_data.csv')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
numerical_features = ['Edad del Niño', 'Ingresos del Hogar (USD)']
categorical_features = ['Sexo del Niño', 'Nivel de Estudios del Padre']

encoded_features = pd.get_dummies(data[categorical_features],
               columns=categorical_features,
               drop_first=True)

data_to_model_standarized = StandardScaler().fit_transform(data[numerical_features])

data_to_model_df = pd.DataFrame(data_to_model_standarized,
                                columns=numerical_features).reset_index()

encoded_features_df = encoded_features.reset_index()

data_to_model = data_to_model_df.merge(encoded_features_df, on='index')
data_to_model = data_to_model.drop('index', axis=1)
data_to_model.head()
```

```python
sum_of_squared_distances = []
K = range(1, 15)  # Ajustar el rango según sea necesario
for k in K:
    km = KMeans(n_clusters=k, random_state=42)
    km = km.fit(data_to_model)
    sum_of_squared_distances.append(km.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(K, sum_of_squared_distances, 'bx-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Sum of Squared Distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
labels = kmeans.fit_predict(data_to_model)

data_to_model['Cluster'] = labels
data['Cluster'] = labels
data.head()
```

```python
for col in numerical_features:
    plt.figure(figsize=(5, 5))
    sns.boxplot(x='Cluster', y=col, data=data, hue='Cluster', palette='pastel')
    plt.title(f'Boxplot de {col} por Cluster')
    plt.show()
```

Dividiendo a los clientes (niños) en cuatro grupos, se logra ver, que hay dos grupos (cluster 0 y 2) que tienen una edad mayor (entre 15 y 14 años) y dos grupos (cluster 1 y 3) con edad mas pequeña (aproximadamente 7 años).

Además, se puede observar un comportamiento similar en la distribución de los ingresos del hogar. Los clusters 0 y 3 cuentan con un ingreso del hogar promedio un poco menor a \$60,000 USD. Los ingresos de los clusters 1 y 2 rondan en promedio entre \$35,000 y \$40,000 USD.

```python
for col in categorical_features:
    sns.countplot(x='Cluster', hue=col, data=data, palette='pastel')
    plt.title(f'Count by {col} by Cluster')
    plt.xlabel('Cluster')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```

Se puede observar que la mayoria de clientes son hombres. En 3 de los cuatro clusters la mayoría de clientes son hombres, solamente en el cluster 1 hay una mayoría de mujeres, por poco.

La mayoría de padres de los niños son "undergraduate", y los padres con "High School" o "Graduate" parecen ser mas o menos iguales en cantidad.

Estas categorías podrían ayudar a la empresa a enfocarse por ejemplo en el cluster 0, ya que se trata de niños "mayores" con un ingreso en casa de alrededor de $60,000 USD, con mayoría hombres, cuyos padres son distribuidos equitativamente entre los grados de estudios. Les serviría enfocarse en este cluster, ya que al tratarse de más ingreso, significa posible ingreso mayor. Ya que son niños más grandes pueden influir en sus padres para gastar más dinero en el juego.

**Opción 4: Tasa de Churn**
   - *Definición*: El porcentaje de clientes que cancelan el servicio durante un período de tiempo determinado.
   - *Fórmula*: $\text{Tasa de Churn} = \frac{\text{Número de Clientes que Cancelaron}}{\text{Número Total de Clientes Activos al Inicio del Periodo}}$

Con el Churn Rate, claramente pueden medir la retención de clientes, ya que es lo que dice explícitamente el índicador. También les sirve de manera implicita para cuantificar su capacidad de generar ingresos a largo plazo, ya que si te tienen clientes, se esperaría que estos consuman dentro de empresa, generando ingresos.

##**Sección 3**

```python
categorical_features = ['Plan de Subscripción','Cluster']
numerical_features = ['Edad del Niño','Número de Sesiones por Semana','Nivel de Satisfacción del Cliente','Ingresos del Hogar (USD)']

data_encoded = pd.get_dummies(data[['ID del Cliente'] + categorical_features], columns=categorical_features, drop_first=True)

df_data_with_encoded = data[['ID del Cliente'] + numerical_features].merge(
    data_encoded, on = 'ID del Cliente'
)

cosa = df_data_with_encoded.groupby("ID del Cliente")[df_data_with_encoded.columns[1:]].mean()
cosa.head()
```

```python
grouped_LTV = pd.DataFrame(data.groupby('ID del Cliente')['Valor del Tiempo de Vida (LTV)'].sum())

todo = grouped_LTV.merge(cosa, on = 'ID del Cliente')
todo.head()
```

```python
X1 = ['Plan de Subscripción_Familiar','Plan de Subscripción_Premium','Cluster_1','Cluster_2','Cluster_3',
      'Edad del Niño','Número de Sesiones por Semana','Nivel de Satisfacción del Cliente','Ingresos del Hogar (USD)']
y1 = ['Valor del Tiempo de Vida (LTV)']

to_remove_x = ['Edad del Niño','Cluster_2','Cluster_1']
X1 = [item for item in X1 if item not in to_remove_x]

X1 = sm.add_constant(todo[X1])

model_1 = sm.OLS(todo[y1], X1)
results = model_1.fit()

print(results.summary())
```

```python
coeficientes = results.params

print("La ecuación de la regresión es: y = " + str(round(coeficientes[1], 4)) + "*Plan de Subscripción_Familiar + " +
      str(round(coeficientes[2], 4)) + "*Plan de Subscripción_Premium + " + str(round(coeficientes[3], 4)) + "*Cluster_3 + \n                                    " +
      str(round(coeficientes[4], 4)) + "*Número de Sesiones por Semana + " + str(round(coeficientes[5], 4)) + "*Nivel de Satisfacción del Cliente + " +
      str(round(coeficientes[6], 4)) + "*Ingresos del Hogar (USD)" + "\n\n")
```

Las variables que tienen impacto sobre el LTV son:
- **Plan de subscripción familiar:** Por cada plan de subscipción familiar se espera un aumento en el LTV de \$711.09.
- **Plan de subscripción premium:** Por cada plan de subscipción familiar se espera un aumento en el LTV de \$282.13.
- **Cluster 3:** Por cada persona que se una al cluster 3 se espera un aumento en el LTV de \$24.37.
- **Número de Sesiones por Semana:** Por cada sesión a la semana se espera un aumento en el LTV de \$17.24.
- **Nivel de satisfacción del cliente:** Por cada unidad de satisfacción del cliente se espera un aumento en el LTV de \$74.43.
- **Ingresos del hogar:** Por cada dólar que genere la familia se espera un aumento en el LTV de \$0.01.

Los coeficientes que más impacto tienen son aquellos cuyos coeficientes son más altos, ya que representan un umpacto más alto en el cálculo del LTV esperado. Tampoco que se puede decir que los ingresos por hogar son despreciables, ya que se trata de una variable con una media mucho mayor a, por ejemplo, nivel de satisfacción. Aunque todas las variables tienen un impacto en el cálculo/predicción del LTV, las que más impacto, de manera directa generan son las que cuentan con un coeficiente más elevado.
