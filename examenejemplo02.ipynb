A continuación se presentan las opciones de KPIs:

### **Opción 1: Tasa de Interacciones del Servicio al Cliente**
   - *Definición*: El número promedio de interacciones del servicio al cliente por cliente.
   - *Fórmula*: $\text{Tasa de Interacciones} = \frac{\text{Interacciones del Servicio al Cliente}}{\text{Clientes Existentes}}$

### **Opción 2: Ingreso Promedio por Cliente (ARPU)**
   - *Definición*: El ingreso promedio generado por cada cliente en un mes.
   - *Fórmula*: $\text{ARPU} = \frac{\text{Ingresos Totales}}{\text{Clientes Existentes}}$

### **Opción 3: Tasa de Crecimiento de Nuevos Clientes**
   - *Definición*: El porcentaje de aumento o disminución en la cantidad de nuevos clientes adquiridos en comparación con el mes anterior.
   - *Fórmula*: $\text{Tasa de Crecimiento de Nuevos Clientes} = \frac{\text{Nuevos Clientes}_{\text{mes actual}} - \text{Nuevos Clientes}_{\text{mes anterior}}}{\text{Nuevos Clientes}_{\text{mes anterior}}}$

### **Opción 4: Tasa de Retorno de Clientes**
   - *Definición*: El porcentaje de clientes existentes que regresan a hacer compras adicionales en un mes dado.
   - *Fórmula*: $\text{Tasa de Retorno de Clientes} = \frac{\text{Clientes que Regresan}}{\text{Clientes Existentes}}$

Opción 4: Tasa de retorno de clientes

Ya que la empresa busca expandir su base de clientes, mientras mantiene un fuerte enfoque en la retención de clientes, se le recomienda utilizar como su KPI la "Tasa de crecimiento de nuevos clientes". Este indicador, se calcula mediante la fórmula: $$\text{Tasa de Retorno de Clientes} = \frac{\text{Clientes que Regresan}}{\text{Clientes Existentes}}$$

El indicador muestra en porcentaje de clientes que regresan a hacer compras dentro del mismo mes de recolección de datos. Ya que la empresa está en crecimeinto se le recomienda ampliamente el enfocarse en este KPI en específico, ya que al no estar todavía establecida es mejor que aumente su base de clientes, de modo que tenga como sobrebibir. Con este KPI se pueden analizar ambas metas de la empresa, tanto el enfoque en la retención de clientes (claramente), como la expansión de su base de clientes, como resultado indirecto, esto porque si se enfocan en que sus clientes regresan, estarían disminuyendo la cantidad de clientes que hacen compras únicas, y por ende aumentando su base.

###**Sección 3**

```python
data.head()
```

```python
x = ['Marketing_Expenditure', 'Web_Traffic', 'Customer_Satisfaction', 'Product_Price', 'Economic_Conditions', 'Promotions']
y = ['Total_Revenue']

to_remove_x = ['Marketing_Expenditure', 'Web_Traffic', 'Customer_Satisfaction', 'Product_Price']

x = [item for item in x if item not in to_remove_x]

X = sm.add_constant(data[x])

model = sm.OLS(data[y], X)
results = model.fit()

print(results.summary())
```

```python
coeficientes = results.params

print("La ecuación de la regresión es: y = " + str(round(coeficientes[1], 8)) + "*Economic_Conditions + " +
      str(round(coeficientes[2], 8)) + "*Promotions + " + str(round(coeficientes[0], 8)) + "\n\n")
```

La regresión lineal nos muestra que las únicas variables que afectan al ingreso total son: condiciones económicas y promociones. Si ambas variables son 0, se espera un ingreso de -\$380,986.05, es decir pérdidas, lo cual es lógico. Por cda unidad de condicion económica que aumente, se espera un aumento de \$498,222.01. De manera similar con las promociones se espera un incremento de \$367,419.80

En el modelo final no hay ningún coeficiente o signo "raro", ya que hace sentido que los ingresos bajen si no hay condiciones económicas o promociones, pero que igualmente el ingreso sube, en medida que estos estén presentes.
